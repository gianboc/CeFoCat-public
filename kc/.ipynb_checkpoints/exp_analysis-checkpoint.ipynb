```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# from matplotlib import 
# plt.rcParams["figure.figsize"] = [16,9]
```

```python
kc = pd.read_csv('./kc_flowfield_new.csv',index_col=False)
```

```python
kc = kc.set_index('case')
```

```python
d = kc[['ppi','por','k','Re','eps_s']].sort_index()
```

```python
d
```

```python
k_b = 1.38e-23
T = 293
nu = 1e-06
mu = 1e-03
rho = 1e03
```

```python
pe_s = np.log10(5)
pe_e = np.log10(500)
Pe = np.logspace(pe_s,pe_e,5).round().astype(int)
```

```python
dt = np.array([d['Re'].values*nu/elem for elem in Pe])
```

```python
len(dt)
```

```python
DT = pd.DataFrame(index=d.index,columns=['Pe=5','Pe=16','Pe=50','Pe=160','Pe=500'] )
# DT['Pe=5'] = dt[0]
for elem,i in zip(DT.columns,range(len(dt))):
    DT[elem] = dt[i]
```

```python
dc = k_b*T/3/np.pi/mu/DT*1e9
```

```python
d.loc['kc_10_95','Re']
```

```python
ms = 20
plt.scatter(d['Re'],dc['Pe=5'],s=ms,facecolors='none',edgecolors='k',label='Pe=5')
plt.scatter(d['Re'],dc['Pe=16'],s=ms,facecolors='none',edgecolors='r',label='Pe=16')
plt.scatter(d['Re'],dc['Pe=50'],s=ms,facecolors='none',edgecolors='b',label='Pe=50')
plt.scatter(d['Re'],dc['Pe=160'],s=ms,facecolors='none',edgecolors='orange',label='Pe=160')
plt.scatter(d['Re'],dc['Pe=500'],s=ms,facecolors='none',edgecolors='g',label='Pe=500')
plt.yscale('log')
plt.xscale('linear')
plt.xlabel('Re number')
plt.ylabel('$d_c$ (nm)')
# plt.yticks([0.5,1,5,10,50,100,500,1000,5000])
plt.xlim(0,0.012)
plt.ylim(0.1,2000)
plt.grid(which='both')
plt.legend()
plt.savefig('dc_vs_Re.png',dpi=600)
```

```python
d.to_csv('data_dp_1e-07.csv')
DT.to_csv('DT_dp_1e-07.csv')
dc.to_csv('dc_dp_1e-07.csv')
```
